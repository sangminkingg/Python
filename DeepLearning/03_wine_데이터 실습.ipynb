```python
# 데이터 불러오기
import pandas as pd
wine = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/사물지능_딥러닝_2022/data/wine.csv")

wine.head()
```

```python
y = wine["type"]
del wine['type']
X = wine

X.shape, y.shape
```

```python
# 신경망 설계
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

model2= Sequential()

model2.add(Dense(units=24, input_dim=12, activation="relu"))

model2.add(Dense(units=48, activation="relu"))
model2.add(Dense(units=24, activation="relu"))

# 회귀인 경우 units는 1로 activation은 linear을 쓰거나 activation을 설정하지 않음
# model9.add(Dense(units=1, activation="linear"))
model2.add(Dense(units=1, activation="sigmoid"))

model2.summary()
```

```python
model2.compile(loss="binary_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
model2.fit(X,y,batch_size=20, epochs=100)
```

```python
#모델 평가
model2.evaluate(X, y)
```

```python
#모델 저장
# 가장 마지막 학습의 모델을 저장
model2.save('/content/drive/MyDrive/Colab Notebooks/사물지능_딥러닝_2022/wine_model.h5')
```

```python
# 신경망 설계
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

model2= Sequential()

model2.add(Dense(units=24, input_dim=12, activation="relu"))

model2.add(Dense(units=48, activation="relu"))
model2.add(Dense(units=24, activation="relu"))

# 회귀인 경우 units는 1로 activation은 linear을 쓰거나 activation을 설정하지 않음
# model9.add(Dense(units=1, activation="linear"))
model2.add(Dense(units=1, activation="sigmoid"))

model2.summary()
```

```python
model2.compile(loss="binary_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
# 학습 중단 : 모델이 더이상 개선되지 않으면 학습을 중단
# 베스트 모델 저장 : 전  epoch보다 좋아지면 저장 
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# 베스트 모델을 저장하는 폴더
model_dir ="/content/drive/MyDrive/Colab Notebooks/사물지능_딥러닝_2022/model"

# 저장할 파일명
model_file = model_dir + "wine_{epoch:02d}-{accuracy:.4f}.hdf5"                                  

mcp = ModelCheckpoint(
    filepath = model_file, # 파일이 저장될 경로
    monitor = 'accuracy', # 정확도를 기준으로 best모델을 판단
    save_best_only = True, # 모델이 개선되었을때만 저장
    verbose = 1 #  진행 내용 출력
    
)

# 학습중단
es = EarlyStopping(
    monitor ="accuracy", # 정확도를 기준으로 모델을 판단
    patience = 5 # 5번의 학습동안 모델이 개선되지 않으면 학습을 중단
)
```

```python
# 6 학습
# 문제데이터, 정답데이터, batch_size, 학습횟수
model2.fit(X, y, batch_size = 100, epochs =100, callbacks=[mcp, es])
```
