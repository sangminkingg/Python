## Age 결측치 채우기

- 결측치가 있는 데이터의 다른 컬럼의 값과 같은 데이터의 평균값을 사용해서 결측치를 채움

- 그러면 어떤 컬럼을 참조할까요?
  - 결측치가 있는 컬럼(Age)와 상관관계가 높은 컬럼 선택(범주형)
  
- 피벗 테이블을 활용

- apply()을 이용하여 전체 데이터의 결측치를 채움
  - 데이터프레임의 데이터를 분리해서 원하는 처리를 수행한 후 다시 병합하는 함수 -> 재구조화 함수 (reconstruct function)

```python
# Age 컬럼과 상관관계가 높은 컬럼찾기
train.corr()
```

```python
# 피벗 테이블 생성
# values : 결측치를 채울 컬럼
# index : 참고할 컬럼 목록 리스트
# aggfunc : 사용할 수학 도구 (평균, 중간값)
pt1 = train.pivot_table(values="Age",
                        index=["Pclass", "Sex"],
                        aggfunc = "mean")

pt1
```

```python
# 피벗 테이블에 접근하는 방법
pt1.loc[1, "male"]
```

```python
import numpy as np
# Age 컬럼의 결측치를 채우는 함수
def fill_age(row):
    # 한 줄 데이터에서 Age 컬럼이 결측치라면
    if np.isnan(row["Age"]):
        # 피벗 테이블을 참조 (같은 Pclass와 Sex인 값을 반환)
        return pt1.loc[row["Pclass"], row["Sex"]]
    # 결측치가 아닌경우
    else :
        return row["Age"]
```

```python
# axis=1 : 한 줄 씩 넘긴다.
# astype("int64") : 나이는 실수가 없으니 정수로 변환
# apply는 pandas or numpy에서 사용자 정의 기능을 적용할때 사용
train["Age"] = train.apply(fill_age, axis=1).astype("int64")
test["Age"] = test.apply(fill_age, axis=1).astype("int64")
```

## 인코딩

```python
# 인코딩을 해주는 컬럼의 순서가 있는지 생각한다. ex)순서가 있다면 라벨인코딩 없다면 원핫인코딩
train.columns  #인코딩을 해줘야하는 컬럼(Sex, Cabin, Embarked, Family_cat, Title)
```

```python
# 인코딩할 컬럼명만 선택
categorical_features = ['Sex', 'Cabin', 'Embarked', 'Family_cat', 'Title']
```

```python
#pd.get_dummies(train['Sex'])
```

```python
# 반복문으로 인코딩하기, prefix = () 인코딩할때 컬럼에서 중복된 키워드가 있을때 쓰는것
for feature_name in categorical_features :
    one_hot = pd.get_dummies(train[feature_name], prefix=feature_name)
    train = pd.concat([train, one_hot], axis=1) # 기존데이터 끝에 one_hot을 붙임
    train.drop(feature_name, axis=1,inplace=True) #기존 글자 컬럼을 삭제
```

```python
train.head()
```

```python
# 테스트 데이터도 동일하게 진행
for feature_name in categorical_features :
    one_hot = pd.get_dummies(test[feature_name], prefix=feature_name)
    test = pd.concat([test, one_hot], axis=1)
    test.drop(feature_name, axis=1,inplace=True)
```

### 모델 최적화를 위해 검증 데이터셋을 만들자

```python
from sklearn.model_selection import train_test_split
```

```python
X_train2,X_val,y_train2,y_val = train_test_split(X_train, y_train,
                                                test_size=0.3,
                                                random_state=719)
```

```python
X_val.shape, y_val.shape
```

```python
X_train2,X_val,y_train2,y_val
```

#### KNN 튜닝

```python
train_acc = [] #훈련데이터의 점수가 들어갈 리스트
test_acc = [] #검증데이터의 점수가 들어갈 리스트

#이웃을 1~19개까지 변경하면서 모델생성
neighbor =range(1, 20)

for n in neighbor :
    #모델학습
    knn_model = KNeighborsClassifier(n_neighbors=n)
    knn_model.fit(X_train2, y_train2)
    
    #훈련데이터에 대한 점수 누적
    train_acc.append(knn_model.score(X_train2, y_train2))
    #테스트데이터에대한 점수 누적
    test_acc.append(knn_model.score(X_val, y_val))
    
import matplotlib.pyplot as plt

plt.plot(neighbor, train_acc, label="Train2")
plt.plot(neighbor, test_acc, label="val")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(neighbor, train_acc, label="Train2")
plt.plot(neighbor, test_acc, label="val")
plt.grid()
plt.xticks(range(1,20))
plt.legend()
plt.show()
```

### 성능개선을 위해 고려해볼 사항
1. 특성선택
2. 데이터 확보
3. 전처리 작업 변경(결측치, 특성공학, 스케일링 등)
4. 모델변경 - 앙상블

# 앙상블 
1. 보팅(voting)
   - 서로 다른 알고르즘의 모델 여러개를 이용해서 투표로 예측결과를 만들어 낸다.\
   - 하드보팅: 모델의 최종 예측결과로 투표
   - 소프트보팅: 모델의 확률결과로 투표
2. 배깅(bagging)
   - 같은 알고리즘 모델 여러개를 이용해서 투표를 진행
   - 여러개의 모델들이 학습할때 다른 의견이 나올 수 있도록 부트스트래핑 작업을 한다.
   - 매번 모델이 학습할때마다 랜덤하게 데이터를 추출하되, 중복된 추출을 허용한다.
   - 랜덤포레스트 : DecisionTree의 배깅모델
3. 부스팅(boosting)
   - 같은 알고리즘 모델 여러개를 이용해서 투표를 진행
   - 이전 모델이 잘못 학습한 결과를 다음모델이 반영해서 학습하도록하는 기법
   - 학습 진행이 순차적으로 일어나기때문에 상대적으로 속도가 느리다.(직렬진행)

### 그리드서치
- 앙상블 모델의 경우 하이퍼파라미터 수가 엄청 많다.
- 각 파라미터를 독립적으로 튜닝하는건 불가능하다.
- 많은 양의 파라미터를 동시에 튜닝하기위한 함수로 그리드서치를 지원한다.

```python
from sklearn.model_selection import GridSearchCV   #CV =  crossval 교차검증 
```

```python
# 탐색;하고 싶은 하이퍼파라미터 조합을 딕셔너리로 생성
param_grid = {
    'max_depth' : [5,10,15,20],
    'n_estimators' :[100,150,200,250],
    'max_features' : [0.5,0.7],
    'min_samples_leaf' : [10,20,25]
}
```

```python
model = RandomForestClassifier(random_state=721)
# 튜닝할 모델, 사용할 파라미터, 교차검증 횟수
grid = GridSearchCV(model,param_grid, cv=3, n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
# 최고 점수 확인
grid.best_score_
```

```python
# 최고 조합으로 학습완료된 모델
best_model = grid.best_estimator_
```

```python
pre = best_model.predict(X_test)
```
