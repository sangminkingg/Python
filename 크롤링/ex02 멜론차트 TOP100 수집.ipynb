```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. req를 활용해서 멜론차트 페이지 정보 요청해서 받아오기
# 2. <Response [406]> = 통신에 실패(요청에서 처리-클라이언트)
# 3. 에러코드가  400번대(클라이언트), 500번대(서버)
req.get("https://www.melon.com/chart/index.htm")  
```

```python
# 브라우저인척 속이는 작업이 필요,  headers 안에 정보를 집어넣어 속인다.
# f12 개발자 도구에서 network탭에서 새로고침 후 원하는 페이지의 type document 에서 앞 부분 클릭 후 나오는 창의 마지막 headers 복사
h ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
```

```python
res = req.get("https://www.melon.com/chart/index.htm",headers=h)
```

```python
soup = bs(res.text,"lxml")
```

###  노래제목 수집

```python
# 노래제목에 해당하는 태그정보를 알아온 다음에
# soup.select() 노래제목만 수집
title = soup.select("div.ellipsis.rank01>span>a")
```

```python
title
```

```python
for i in title:
    print(i.text)
```

# 가수 100명 수집

```python
# 가수명에 해당하는 태그정보를 알아온 다음에
# soup.select()가수 이름만 수집
#클래스값 사이에 공백이 있다는 의미는 클래스가 복수개
#공백을 지우고 .을 통해서 크르래스를 이어준다
sname = soup.select("div.ellipsis.rank02>a")
```

```python
for i in sname:
    print(i.text)
```

```python
# 수집한 데이터를 사용자가 사용할 수 있게 변형(태그를 삭제, 글자만 저장)
# 순수한 텍스트 정보만 새로 제작
singerList = [] # 순수한 가수이름정보
titleList = [] # 순수한 노래제목정보

#for 문 사용
for i in range(len(singer)):
    singerList.append(singer[i].text)
    titleList.append(title[i].text)
```

```python
titleList
```

```python
# 변형된 데이터를 표형식으로 제작
import pandas as pd
```

```python
dic = {"가수": singerList, "노래제목":titleList}
```

```python
# 리스트 : 같은데이터를 다룰때 씀
# 딕셔너리 : 다른 데이터를 다룰때 슴
melon = pd.DataFrame(dic)
```

```python
# 표를 파일로 저장
# 나 melon 표를 csv 형태로 저장 해줘
# 한글을 저장할때는 encoding 필수!
# encoding 팁 1부터 안될때 차례대로 적용
## 1. euc-kr 적용안되는 컴퓨터있을때
## 2. utf-8
## 3. utf-8-sig
## 4. "" 될때 안될때 있음
melon.to_csv("멜론차트.csv",encoding='euc-kr')
```
