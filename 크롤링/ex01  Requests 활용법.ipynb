```python
# import 작업
# requests = 브라우저의 역할을 대신(요청, 응답정보를 처리)
import requests as req
```

```python
# 1. 네이버 url을 통해서 사이트를 요청
# req야 네이버 페이지좀 가져와줘!
# <Response [200]> 통신에 성공했습니다
res = req.get("http://www.naver.com")
```

```python
# 2. 응답받은 페이지에서 원하는 정보를 추출(카페, 블로그...)
res.text
```

# 컴퓨터가 알아먹을 수 있는 언어로 변형
## beautifulSoup 라이브러리
### 응답받은 텍스트 데이터를 html데이터로 변형

```python
from bs4 import BeautifulSoup as bs     #BeautifulstoneSoup 과 헷갈리지 말것!! 
```

```python
# bs아 내가 가지고있는 글자정보를 html정보로 바꿔줘~
# bs("내가 가지고있는 정보", "파싱방법") 
soup = bs(res.text, "lxml")   
soup
```

```python
#네이버 페이지에서 카페, 블로그, 지식인... 글자를 수집해!
#데이터에서 a태그만 수집해줘~   # html언어에서 ( . ) = class (#) = id 
data = soup.select("a.nav")
# 리스트로 나오기때문에 data[0] 을 출력하면 메일이 출력됌
```

```python
for i in data:
    print(i.text)
```

# 뉴스 타이틀 수집

```python
ress =req.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%9E%A5%EB%A7%88")
```

```python
ress.text
```

```python
soup2 = bs(ress.text, "lxml")
```

```python
title = soup2.select("a.news_tit")
```

```python
for i in title:
    print(i.text)
```
